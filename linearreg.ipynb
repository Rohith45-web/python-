```python
## atributes of numpy
arr=np.array([[1,2,3],[4,6,5]])
print('array:\n',arr)
print('shape',arr.shape)
print('number of dimensins:',arr.size)
print('Datatype:',arr.dtype)

```

```python
## numpy vectorized operation
arr1=np.array([1,2,3,4,5])
arr2=np.array([2,4,6,7,8])
print('addition:',arr1+arr2)
print('substraction:',arr2-arr1)
print('multiplication:',arr1*arr2)
print('division:',arr1/arr2)
```

```python
import tkinter as tk
from tkinter import messagebox

# -------------------------------
# Sample Rule-Based Diagnosis Logic
# (Replace this with your own rules)
# -------------------------------
def diagnose(symptoms):
    symptoms = symptoms.lower()

    if "fever" in symptoms and "cough" in symptoms:
        return "Viral Fever"
    elif "headache" in symptoms and "nausea" in symptoms:
        return "Migraine"
    elif "chest pain" in symptoms or "breathlessness" in symptoms:
        return "Possible Heart Issue (Consult Doctor)"
    elif "vomit" in symptoms and "loose motion" in symptoms:
        return "Food Poisoning"
    else:
        return "No matching condition found. Add more rules."

# -------------------------------
# UI Function
# -------------------------------
def diagnose_user():
    user_symptoms = entry.get()

    if user_symptoms.strip() == "":
        messagebox.showwarning("Input error", "Please enter symptoms!")
        return

    result = diagnose(user_symptoms)
    result_label.config(text="Diagnosis: " + result)


# -------------------------------
# Tkinter GUI
# -------------------------------
window = tk.Tk()
window.title("Health Diagnosis System")
window.geometry("450x300")
window.config(bg="#f2f2f2")

title = tk.Label(window, text="Health Diagnosis System", font=("Arial", 18, "bold"), bg="#f2f2f2")
title.pack(pady=10)

entry = tk.Entry(window, font=("Arial", 14), width=35)
entry.pack(pady=10)
entry.insert(0, "Enter symptoms (e.g., fever, cough)")

btn = tk.Button(window, text="Diagnose", font=("Arial", 14), bg="#4CAF50", fg="white",
                command=diagnose_user)
btn.pack(pady=10)

result_label = tk.Label(window, text="Diagnosis: ", font=("Arial", 14), bg="#f2f2f2")
result_label.pack(pady=20)

window.mainloop()
```
