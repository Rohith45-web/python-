```python
## independent and dependent features
x=df.drop('total_bill',axis=1)
y=df['total_bill']
```

```python
##train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le1=LabelEncoder()
le2=LabelEncoder()
le3=LabelEncoder()
```

```python
x_train['sex']=le1.fit_transform(x_train['sex'])
x_train['smoker']=le2.fit_transform(x_train['smoker'])  
x_train['time']=le3.fit_transform(x_train['time'])
```

```python
x_train.head()
```

```python
x_test['sex']=le1.fit_transform(x_test  ['sex'])
x_test['smoker']=le2.fit_transform(x_test['smoker'])    
x_test['time']=le3.fit_transform(x_test['time'])
```

```python
## onehot encoding
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[1,3])],remainder='passthrough')
x_train=np.array(ct.fit_transform(x_train))
x_test=np.array(ct.transform(x_test))
```

```python
from sklearn.svm import SVR
svr=SVR()
svr.fit(x_train,y_train)
y_pred=svr.predict(x_test)
from sklearn.metrics import r2_score,mean_absolute_error
r2_score(y_test,y_pred)
```

```python
x_test
```

```python
print(mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={'C':[0.1,1,10,100],'gamma':[1,0.1,0.01,0.001],'kernel':['rbf']}
grid=GridSearchCV(SVR(),param_grid,refit=True,verbose=2)
grid.fit(x_train,y_train)
y_pred=grid.predict(x_test)
r2_score(y_test,y_pred)

```

```python
grid.best_params_
```

```python
grid_pred=grid.predict(x_test)
r2_score(y_test,grid_pred)
print(mean_absolute_error(y_test,grid_pred))
```
